### Crawl Data: https://vinmart.co/danh-muc
#### Danh mục chọn: https://vinmart.co/cham-soc-ve-sinh-nha-cua-6502/

#### Import thư viện 

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support.ui import Select
```

### Link sản phẩm 10 page  đầu tiên

```python
service = Service(executable_path="C:\chromedriver\chromedriver.exe")
driver = webdriver.Chrome(service=service)
#Tạo list chứa các link sản phẩm
product_links = []

for i in range(1,11):
    driver.get("https://vinmart.co/cham-soc-ve-sinh-nha-cua-6502/?trang="+str(i))
    links = driver.find_elements("xpath","//div[@class='x-box x-shadow']/a")
    print("Links sản phẩm trang "+str(i)+": ")
    for link in links:
        product_links.append(link.get_attribute("href"))
        print(link.get_attribute("href"))
```

### Tên sản phẩm (String)

```python
# Tạo biến product_names để lưu tên sản phẩm (string)
product_names = []
for link in product_links:
    driver.get(link)
    product_name = driver.find_element("xpath","//span[@class='product-name']")
    product_names.append(product_name.text)
    print(product_name.text)
```

### Cửa hàng (string)

```python
# Tạo biến stroes_names để lưu tên cửa hàng
stores_names = []
for link in product_links:
    driver.get(link)
    link_store = driver.find_elements("xpath","//div[@class='product-info pt-2']/a")
    store = driver.find_element("xpath","//div[@class='product-info pt-2']")
    stores_names.append(store.text)
    print(store.text)
    print(link_store[0].get_attribute("href"))
```

```python
# Tên cửa hàng thành (string) và loại bỏ từ "\nXem Shop"
stores_names = [str(i) for i in stores_names]
stores_names = [i.replace("\nXem Shop","") for i in stores_names]
print(stores_names)
```

### Crawl  Giá bán (float)

```python
# Tạo biến product_prices để lưu giá sản phẩm
product_prices = []
for link in product_links:
    driver.get(link)
    product_price = driver.find_element("xpath","//span[@class='special-price']")
    product_prices.append(product_price.text)
    print(product_price.text)
    
# Lưu giá bán dưới dạng float và bỏ ký tự đặc biệt
product_prices_float = []
for price in product_prices:
    product_prices_float.append(float(price.replace("₫","").replace(".","")))
```

```python
# Giá bán float
print(product_prices_float)
```

### Trạng thái (nếu có) (string)

```python
# Tạo biến product_status để lưu trạng thái sản phẩm (string)
product_status = []
for link in product_links:
    driver.get(link)
    status = driver.find_element("xpath","//div[@class='product-info']")
    product_status.append(status.text)
    print(status.text)
```

### Kho hàng (nếu có) (string)


```python
# Tạo biến product_stock để lưu kho hàng (string)
product_stock = []
for link in product_links:
    driver.get(link)
    stock = driver.find_element("xpath","//div[@class='product-info']/span")
    # Kiểm tra xem sản phẩm có kho hàng hay không do nếu không có kho hàng thì thẻ span sẽ cào giá tiền về
    if "₫" in stock.text:
        product_stock.append("Mặt hàng này không có kho hàng")
        print("Mặt hàng này không có kho hàng")
    else:
        product_stock.append(stock.text)
        print("Kho hàng:",stock.text)
```

### Mô tả sản phẩm

```python
# Tạo biến product_descriptions để lưu mô tả sản phẩm
product_descriptions = []
for link in product_links:
    driver.get(link)
    descriptions = driver.find_element("xpath","//div[@class='tab-pane fade show active']")
    product_descriptions.append(descriptions.text)
    print(descriptions.text)   
```

### Lưu dữ liệu vào file txt


```python
# Tạo file
logfile=open(r"C:\Users\ADMIN\OneDrive\Desktop\Studies\NMKHDL\CrawlData\datavinmart.txt",'w')
# Ghi dữ liệu vào file
with open(r"C:\Users\ADMIN\OneDrive\Desktop\Studies\NMKHDL\CrawlData\datavinmart.txt","w",encoding="utf-8") as f:
    f.write("Data Vinmart: \n")
    for i in range(1,11):
        driver.get("https://vinmart.co/cham-soc-ve-sinh-nha-cua-6502/?trang="+str(i)+"\n")
        links = driver.find_elements("xpath","//div[@class='x-box x-shadow']/a")
        f.write("Links sản phẩm trang "+str(i)+": \n")
        for link in links:
            f.write(link.get_attribute("href")+"\n")
            for i in range(len(product_links)):
                f.write("Tên sản phẩm: "+product_names[i]+"\n")
                f.write("Tên cửa hàng: "+stores_names[i]+"\n")
                f.write("Giá sản phẩm: "+product_prices[i]+"\n")
                f.write("Trạng thái sản phẩm: "+product_status[i]+"\n")
                f.write("Kho hàng: "+product_stock[i]+"\n")
                f.write("Mô tả sản phẩm: "+product_descriptions[i]+"\n")
                f.write("\n")
logfile.close()
```
